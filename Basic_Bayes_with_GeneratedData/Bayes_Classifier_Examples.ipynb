# Dataset A

```python
mu1 = np.array([1,1])
cov1 = np.array([[1,0],[0,1]])
mu2 = np.array([4,4])
cov2 = np.array([[1,0],[0,1]])

mu= [mu1,mu2 ]
cov =[cov1, cov2]
n_samples= [60000,140000]
```

```python
dataA = BC.SampleData(mu,cov, n_samples)
dataA.run_all_analysis()
```

```python
dataA.make_plot(plot_type='seaborn', sort_type='raw', title='Samples from Dataset A',mean_lines=True, show_boundary=False)

```

```python
dataA.make_plot(plot_type='seaborn', sort_type='sorted', title='Samples from Dataset A - Classified by Case I Discriminant', show_boundary=True)

```

```python
dataA.make_plot(plot_type='seaborn', sort_type='euclidean', title='Dataset A - Classified by Euclidean Distance', mean_lines=False, show_boundary=True)


```

```python
dataA.summary_stats()
```
